# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
marks = int(input("Enter the current temperature: "))

if marks >= 90:
    print (f'''You got A grade and your marks is, {marks}''')

elif marks >= 80:
    print (f'''You got B grade and your marks is, {marks}''')

elif marks >= 70:
    print (f'''You got C grade and your marks is, {marks}''')

elif marks >= 60:
    print (f'''You got D grade and your marks is, {marks}''')

else:
    print (f'''You are fail and your marks is, {marks}''')
```

# User se day number (1–7) lo:

# 7 → Sunday

```python
day = int(input("Enter the number to get the day: "))

if day == 1:
    print ("It's Monday")

elif day == 2:
    print ("It's Tuesday")
    
elif day == 3:
    print ("It's Wednessday")
    
elif day == 4:
    print ("It's Thursday")
    
elif day == 5:
    print ("It's Friday")
    
elif day == 6:
    print ("It's Saturday")
    
elif day == 7:
    print ("It's Sunday")

else:
    print ("Out of days")
    
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(input("Enter the number of month to find the weather"))

if month == 12:
    print (f"""Current weather of month {month} is Winter""")

elif month <= 2:
    print (f"""Current weather of month {month} is Winter""")

elif month <= 5:
    print (f"""Current weather of month {month} is Spring""")

elif month <= 8:
    print (f"""Current weather of month {month} is Summer""")
    
elif month <= 11:
    print (f"""Current weather of month {month} is Autumn""")

else:
    print ("You are out of Month")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
operator =input("Enter the operator +, -, *, /:" )

if operator == "+":
    print ("Addition")

elif operator == "-":
    print ("Substraction")

elif operator == "*":
    print ("Multiplication")

elif operator == "/":
    print ("Division")

else:
    print ("Invalid operator")
```
